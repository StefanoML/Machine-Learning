# USEFUL PLOTS FOR ML

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

np.random.seed(42)

age = np.random.randint(22, 60, 100)
experience = np.clip(age - np.random.randint(18, 25, 100), 0, None)
salary = 30000 + (experience * 2000) + np.random.randint(-5000, 5000, 100)
department = np.random.choice(['Engineering', 'Marketing', 'Sales', 'HR'], 100)

df = pd.DataFrame({
    'age': age,
    'experience': experience,
    'salary': salary,
    'department': department
})

df.head()

epochs = range(1, 101)
loss = [100 * 0.95**i for i in epochs]  # Simulated decaying loss curve
```

## 1. Histogram
**Description:** Displays the distribution of a single numerical variable by dividing values into bins and counting how many data points fall in each bin.

**When to use:**
- To understand the distribution of a feature (is it normal? skewed? bimodal?)
- To spot outliers
- To check if scaling or transformation is needed before training a model

### Matplotlib

```python
# --- BASIC (minimum required) ---
plt.hist(df['salary'], bins=20)
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Frequency')


# --- EXTRAS (add any of these as needed) ---
# plt.figure(figsize=(8, 5))          # Custom figure size, goes at the start
# plt.hist(..., color='steelblue',    # Bar color
#             edgecolor='black',    # Bar border color
#             orientation='vertical') # 'horizontal' for sideways
plt.grid(axis='y', alpha=0.5)       # 'x', 'y', or 'both' | alpha = transparency
# plt.xticks(range(30000, 120000, 10000), rotation=45)  # Custom x ticks | or pass a list
plt.show()
plt.close()  # Closes the figure and prevents settings bleeding into next plot
```

### Seaborn

```python
# --- BASIC ---
sns.histplot(df['salary'], bins=20)
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Frequency')


# --- EXTRAS ---
# sns.histplot(..., kde=True,                   # Adds a smooth density curve on top
#                   color='steelblue',          # Bar color
#                   edgecolor='black')          # Bar border
sns.set_style('darkgrid')                    # Overall style: 'white', 'whitegrid', 'ticks'
plt.show()
plt.close()  # Closes the figure and prevents settings bleeding into next plot
```

### Pandas

```python
# --- BASIC ---
df['salary'].hist(bins=20)
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Frequency')

# --- EXTRAS ---
# df['salary'].hist(bins=20,
#                   by='department')  # Creates separate histograms grouped by a category
# df.hist(bins=20)                    # Plots histograms for ALL numerical columns at once
plt.show()
plt.close()  # Closes the figure and prevents settings bleeding into next plot
```

## 2. Scatter Plot
**Description:** Displays the relationship between two numerical variables as individual points on a graph.

**When to use:**
- To visualize the relationship between two features
- To spot correlations (positive, negative, or none)
- To identify outliers
- To evaluate model predictions (predicted vs real values)

### Matplotlib

```python
# --- BASIC ---
plt.scatter(df['experience'], df['salary'])
plt.title('Experience vs Salary')
plt.xlabel('Experience (years)')
plt.ylabel('Salary')


# --- EXTRAS ---
# plt.figure(figsize=(8, 5))                    # Always before plt.scatter()
# plt.scatter(..., color='steelblue',           # Point color
#                  alpha=0.5,                   # Transparency, useful when points overlap
#                  s=50,                        # Point size
#                  marker='o')                  # Point shape: 'o', 'x', '+', '^', 's'
# plt.colorbar()                                # Adds a color scale if using c= parameter
# plt.scatter(..., c=df['age'],                 # Color points by a third variable
#                  cmap='viridis')              # Color map: 'viridis', 'plasma', 'coolwarm'
# plt.grid(True)                                # Add grid
plt.show()
plt.close()
```

### Seaborn

```python
# --- BASIC ---
sns.scatterplot(data=df, x='experience', y='salary')
plt.title('Experience vs Salary')
plt.xlabel('Experience (years)')
plt.ylabel('Salary')


# --- EXTRAS (seaborn specific) ---
# sns.scatterplot(..., hue='department',   # Color points by category
#                      size='age',         # Scale point size by a variable
#                      style='department') # Different marker per category

plt.show()
plt.close()
```

### Pandas

## 5. Heatmap
**Description:** Displays a matrix of values as a color-coded grid, where color intensity represents the value magnitude.

**When to use:**
- To visualize correlations between features
- To visualize similarity matrices in recommendation systems
- To spot multicollinearity between features before training

### Matplotlib
*Note: Matplotlib heatmaps are rarely used in practice. Seaborn is the standard. Documented here for completeness.*

```python
# --- BASIC ---
corr_matrix = df[['age', 'experience', 'salary']].corr()
plt.imshow(corr_matrix, cmap='coolwarm')
plt.colorbar()
plt.title('Correlation Heatmap')
plt.xticks(range(len(corr_matrix.columns)), corr_matrix.columns)
plt.yticks(range(len(corr_matrix.columns)), corr_matrix.columns)

# --- EXTRAS ---
# plt.figure(figsize=(8, 6))                    # Always before plt.imshow()
# plt.imshow(..., cmap='viridis')               # Colormaps: 'coolwarm', 'viridis', 'plasma'
# plt.clim(-1, 1)                               # Set color scale limits, useful for correlations

plt.show()
plt.close()
```

### Seaborn

```python
# --- BASIC ---
corr_matrix = df[['age', 'experience', 'salary']].corr()
sns.heatmap(corr_matrix)
plt.title('Correlation Heatmap')

# --- EXTRAS (seaborn specific) ---
# sns.heatmap(corr_matrix,
#             annot=True,                       # Shows the actual values inside each cell
#             fmt='.2f',                        # Format of the values: '.2f' = 2 decimal places
#             cmap='coolwarm',                  # Colormap: 'coolwarm', 'viridis', 'plasma'
#             vmin=-1, vmax=1,                  # Color scale limits, important for correlations
#             linewidths=0.5,                   # Adds grid lines between cells
#             square=True)                      # Forces cells to be square shaped

plt.show()
plt.close()
```

## 6. Bar Chart
**Description:** Displays categorical data as rectangular bars where the height represents the value.

**When to use:**
- To compare values across categories
- To show counts or frequencies per category
- To display feature importance in ML models
- To show average values per group

### Matplotlib

```python
# --- BASIC ---
dept_salary = df.groupby('department')['salary'].mean()
plt.bar(dept_salary.index, dept_salary.values)
plt.title('Average Salary by Department')
plt.xlabel('Department')
plt.ylabel('Average Salary')

# --- EXTRAS ---
# plt.figure(figsize=(8, 5))                        # Always before plt.bar()
# plt.bar(..., color='steelblue',                   # Bar color
#              edgecolor='black',                   # Bar border color
#              width=0.5)                           # Bar width, default is 0.8
# plt.bar(..., orientation='horizontal')            # Use plt.barh() for horizontal bars
# plt.xticks(rotation=45)                           # Rotate x labels to avoid overlap
# plt.grid(axis='y')                                # 'x', 'y', or 'both'

plt.show()
plt.close()
```

### Seaborn

```python
# --- BASIC ---
sns.barplot(data=df, x='department', y='salary')
plt.title('Average Salary by Department')
plt.xlabel('Department')
plt.ylabel('Average Salary')

# --- EXTRAS (seaborn specific) ---
# sns.barplot(..., hue='department',            # Color bars by category
#                  palette='viridis',           # Color palette: 'viridis', 'Set2', 'coolwarm'
#                  errorbar=None,               # Removes confidence interval lines
#                  order=['Engineering',        # Custom order of categories
#                         'Marketing',
#                         'Sales', 'HR'])
# plt.xticks(rotation=45)                       # Rotate x labels to avoid overlap

plt.show()
plt.close()
```
