```python
import pandas as pd
import numpy as np
```

## Finding Matches

```python
data = {
    'name': ['Alice', 'Bob', 'Carol', 'David', 'Emma'],
    'city': ['Stockholm', 'Gothenburg', 'Malmö', 'Uppsala', 'Gothenburg'],
    'age': [34, 28, 45, 31, 26]
}

df = pd.DataFrame(data)
print(df)
```

```python
match = df[df["city"]=="Gothenburg"]
if match.empty:
    print("no match found")
else:
    for i in match.index:
        matches = match.index.tolist()
    print(f"Matching Indexes: {matches}")
```

```python
match = df[df["name"]== "Zara"]
if match.empty:
    print ("No match found")
else:
    matches = match.index.tolist()  
    print(f"Matches found: {matches}")
```

## Sorting

```python
scores = np.array([0.3, 0.8, 0.1, 0.9, 0.5])
sorted_indices = np.argsort(scores)
print(sorted_indices)
```

### invert the sorting

```python
sorted_indices = np.argsort(scores)[::-1]
print(sorted_indices)
```

### Limit the output

```python
sorted_indices = np.argsort(scores)[::-1][:3] #From 0 to 3 - 3 not included
print(sorted_indices)
```

# Combined matching and sorting

```python
data2 = {
    'name': ['Alice', 'Bob', 'Carol', 'David', 'Emma'],
    'score': [0.3, 0.8, 0.1, 0.9, 0.5]
}

df2 = pd.DataFrame(data2)
print(df2)
```

```python
match =df2[df2["name"]=="Carol"]
if match.empty:
    print("No match found")
else:
    matches = match.index.tolist()
print(f"Matches found: {matches}")
print(f"Carol's score = {df2["score"][matches].values[0]}")
```

```python
# Getting a value by index position
df2['score'].iloc[2]      # by position number
df2['score'].loc[2]       # by index label (same thing when index is default)
```

## Argsort

```python
scores = np.array([0.8, 0.1, 0.95, 0.3, 0.5])

sorted_scores = np.argsort(scores)[::-1]
print(sorted_scores)
print(scores[sorted_scores])
```

```python
movies = np.array(["Inception", "Titanic", "The Matrix", "Interstellar", "Parasite"])
print(movies[sorted_scores])
```

```python
scores = pd.Series([0.3, 0.9, 0.1, 0.75, 0.5])
movie_titles = pd.Series(["Inception", "Titanic", "The Matrix", "Interstellar", "Parasite"])
sorted_sc = scores.argsort().values[::-1]
print(movie_titles[sorted_sc])
```

```python
scores = pd.Series([0.45, 0.88, 0.23, 0.91, 0.67])
titles = pd.Series(["Alien", "Gladiator", "Frozen", "Mad Max", "Dunkirk"])

sorted_index = scores.argsort().values[::-1][:3]
print(titles[sorted_index].values)
```
