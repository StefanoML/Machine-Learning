```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

data = {
    "title": ["Jaws", "Dunkirk", "Parasite", "Her", "Gravity"],
    "tags": ["shark ocean thriller horror suspense", 
             "war action history soldiers beach",
             "thriller drama class family dark comedy",
             "romance sci-fi artificial intelligence drama",
             "sci-fi thriller space survival disaster"]
}

df = pd.DataFrame(data)

n = 3

#1 Find index
idx = df[df["title"]== "Gravity"].index[0]

#2 Fit matrix
tfidf = TfidfVectorizer(stop_words="english")
tfidf_matrix = tfidf.fit_transform(df["tags"])

#3 Retrieve vector
vector = tfidf_matrix[idx].toarray()

#4 Cosine Similarity
similarity = cosine_similarity(vector, tfidf_matrix)
sim = similarity[0]

#5 Sort indexes by similarity
sorted_indexes = sim.argsort()[::-1]
top3 = sorted_indexes[1:n+1]

#6 Retrieve the titles
matches = df["title"].iloc[top3].values
print(matches)
```
