```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
documents = [
    "the cat sat on the mat",
    "the dog sat on the log",
    "the cat chased the dog",
    "the fish swam in the sea",
    "the sea is deep and dark"
]
```

```python
df_docs = pd.DataFrame({'text': documents})
print(df_docs)
```

```python
vectorizer = TfidfVectorizer(stop_words="english")
tfidf_matrix = vectorizer.fit_transform(documents)

print(tfidf_matrix.shape)

feature_names = vectorizer.get_feature_names_out()
df_tfidf = pd.DataFrame(tfidf_matrix.toarray(), columns=feature_names)
print(df_tfidf)
```

```python
similarity_matrix = cosine_similarity(tfidf_matrix)
df_similarity = pd.DataFrame(similarity_matrix)
print(df_similarity)
```

```python
row = tfidf_matrix[0].toarray()
print(row)
print(row.shape)
```

```python
def get_document_vector(index):
    row = tfidf_matrix[index].toarray()
    return row

print(get_document_vector(0))
print(get_document_vector(3))
```

```python
def find_similar (index, tfidf_matrix):
    row = get_document_vector(index)
    similarity_matrix= cosine_similarity(row, tfidf_matrix)

    return similarity_matrix

print(find_similar(3,tfidf_matrix ))
```
