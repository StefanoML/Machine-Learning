```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

data = {
    "title": ["Inception", "Titanic", "The Matrix", "Interstellar", "Alien"],
    "tags": ["sci-fi thriller dream heist", "romance disaster ship tragedy", 
             "sci-fi action simulation hacker", "sci-fi space time travel drama", 
             "sci-fi horror alien space thriller"]
}

n=3
df = pd.DataFrame(data)
```

```python
#Build a complete TF-IDF pipeline on the tags column and print the similarity scores between "The Matrix" and all other movies. 

#1 find index
idx = df[df["title"] == "The Matrix"].index[0]
print(idx)
```

```python
#2 fit the tag matrix
tfdif = TfidfVectorizer(stop_words="english")
tfdif_matrix = tfdif.fit_transform(df["tags"])
print(tfdif_matrix.shape)
```

```python
#3 Retrieve the vetor of the movie from the matrix
vector = tfdif_matrix[idx].toarray()
```

```python
#4 Calculate cosine similarity between vector and matrix
similarity = cosine_similarity(vector,tfdif_matrix)
sim = similarity[0]
print(sim)
```

```python
#5 Sort indexes by similarity
sorted_indexes = sim.argsort()[::-1]
top3 = sorted_indexes[1:n+1]
print(top3)
```

```python
#6 Retrieve the titles

matches = df["title"].iloc[top3]
print(matches)
```
